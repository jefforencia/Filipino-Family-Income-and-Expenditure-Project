### 2. Impact of Education on Income
  Do household heads with higher education levels earn more?

```python
# Creating new column for Monthly Income
monthly_income = df['Monthly Income'] = df['Total Household Income'] // 12
```

```python
df.head(2)
```

```python
# Total Household Income, Monthly Income and Main Source of Income
df[['Total Household Income', 'Monthly Income', 'Main Source of Income']].head(20)
```

```python
df['Household Head Highest Grade Completed'].value_counts()
```

```python
# Filtering Household Head Highest Grade Completed at Grade School only.
grade_filter = (df['Household Head Highest Grade Completed'] == 'No Grade Completed') | (df['Household Head Highest Grade Completed'] == 'Preschool') | (df['Household Head Highest Grade Completed'] == 'Grade 1') | (df['Household Head Highest Grade Completed'] == 'Grade 2') | (df['Household Head Highest Grade Completed'] == 'Grade 3') | (df['Household Head Highest Grade Completed'] == 'Grade 4') | (df['Household Head Highest Grade Completed'] == 'Grade 5') | (df['Household Head Highest Grade Completed'] == 'Grade 6')
grade_filter_df = df[grade_filter]
grade_filter_df.groupby('Household Head Highest Grade Completed')['Monthly Income'].max()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="Household Head Highest Grade Completed", y="Monthly Income", data = grade_filter_df)
plt.title("Household Income by Education Level")
plt.ylabel("Monthly Income")
plt.show()
```

```python
# Filtering Household Head Highest Grade Completed at High School only.

high_school_filter = (df['Household Head Highest Grade Completed'] == 'Elementary Graduate') | (df['Household Head Highest Grade Completed'] == 'First Year High School') | (df['Household Head Highest Grade Completed'] == 'Second Year High School') | (df['Household Head Highest Grade Completed'] == 'Third Year High School') | (df['Household Head Highest Grade Completed'] == 'Second Year High School') | (df['Household Head Highest Grade Completed'] == 'Third Year High School') | (df['Household Head Highest Grade Completed'] == 'High School Graduate')
high_school_filter_df = df[high_school_filter]
high_school_filter_df.groupby('Household Head Highest Grade Completed')['Monthly Income'].max()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="Household Head Highest Grade Completed", y="Monthly Income", data = high_school_filter_df)
plt.title("Household Income by Education Level")
plt.ylabel("Monthly Income")
plt.show()
```

```python
# Filtering Household Head Highest Grade Completed with Degree.

with_degree_filter = (df['Household Head Highest Grade Completed'] == 'Law Programs') | (df['Household Head Highest Grade Completed'] == 'Veterinary Programs') | (df['Household Head Highest Grade Completed'] == 'Computing/Information Technology Programs') | (df['Household Head Highest Grade Completed'] == 'Teacher Training and Education Sciences Programs') | (df['Household Head Highest Grade Completed'] == 'Health Programs') | (df['Household Head Highest Grade Completed'] == 'Post Baccalaureate')
with_degree_filter_df = df[with_degree_filter]
with_degree_filter_df.groupby('Household Head Highest Grade Completed')['Monthly Income'].max()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(y="Household Head Highest Grade Completed", x="Monthly Income", data = with_degree_filter_df)
# plt.xticks(rotation=90)
plt.title("Household Income by Education Level")
plt.ylabel("Household Head Highest Grade Completed")
plt.show()
```
